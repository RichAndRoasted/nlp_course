```python
import nltk
from nltk.tokenize import word_tokenize
from collections import Counter
from nltk.corpus import stopwords
from datasets import load_dataset
nltk.download('stopwords')
stopwords_ru = stopwords.words('russian')
```

```python
data = load_dataset('Davlan/sib200', 'rus_Cyrl')
nltk.download('stopwords')
stopwords_ru = stopwords.words('russian')
nltk.download('punkt_tab')
```

```python
data
```

```python
def tokenize(dataset):
    vocab = set()
    for i, sentence in enumerate(dataset):
        words = word_tokenize(sentence['text'], language='russian')
        words = [word.lower() for word in words if word.isalpha() and not word.lower() in stopwords_ru]
        vocab.update(words)
    return vocab
```

```python
train_vocab = tokenize(data['test'])
validation_vocab = tokenize(data['validation'])
test_vocab = tokenize(data['test'])
```
